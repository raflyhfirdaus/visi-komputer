```python
import cv2
import matplotlib.pyplot as plt

# algoritma menggunakan metode Global Thresholding
img = cv2.imread('bhayangkara-fc.png', 0)

# Hitungan threshold.
# Perhatikan nilai ambang batas bawah dan atas dari tiap fungsi
# yang diberikan
ret, thresh1 = cv2.threshold(img, 50, 220, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 50, 220, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 50, 220, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 50, 220, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 50, 220, cv2.THRESH_TOZERO_INV)

# menampilkan hasil
titles = ['Gambar asli', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

# menampilkan beberapa gambar sekaligus
plt.figure(figsize=(10, 8))
for i in range(6):
    # 3 baris, 2 kolom
    plt.subplot(3, 2, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```
